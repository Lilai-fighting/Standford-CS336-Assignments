### Section 2.1: The Unicode Standard

```python
chr(0)
```

It returns NULL character (U+0000).

```python
print(chr(0))
```

String representation is a visible escape sequence (转义形式), while printed representation is invisible control character (控制字符).

```python
"this is a test" + chr(0) + "string"
```

```python
print("this is a test" + chr(0) + "string")
```

When the NULL character occurs in text, it is stored as part of the string but usually appears invisible when printed.

### Section 2.2: Unicode Encodings

UTF-8 is usually preferred because it is compact for ASCII-heavy text and represents text as a simple byte sequence without the extra null bytes or surrogate-pair complications of UTF-16/UTF-32. This makes it more efficient and convenient for byte-level tokenization.

```python
def decode_utf8_bytes_to_str_wrong(bytestring: bytes):
    return "".join([bytes([b]).decode("utf-8") for b in bytestring])
```

```python
"é".encode("utf-8")
```

```python
decode_utf8_bytes_to_str_wrong("é".encode("utf-8"))
```

For example, b'\xc3\xa9', which is "é".encode("utf-8"), is decoded incorrectly because the function tries to decode each byte separately. "\xc3" is not a UTF-8 character.

```python
b'\xc3\x28'.decode("utf-8")
```

An example is b'\xc3\x28', because 0xc3 starts a two-byte UTF-8 sequence but 0x28 is not a valid continuation byte, so the sequence cannot be decoded as Unicode (Continuation byte ranges from 0x80 to 0xBF).
